```python
def plotf(f):
    a = -10
    b = 11
    out = [f(x) for x in range(a, b)]
    plt.grid()
    plt.plot([z for z in range(a, b)], out)
```

```python
plotf(lambda x: x**2)
```

```python
def relu(x): return max(0, x)
```

```python
plotf(lambda x: relu(x) * 5 * 2*x * relu(x))
```
